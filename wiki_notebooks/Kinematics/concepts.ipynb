# Distance and Displacement

Now that we have learned about linear motion, let us discuss two terms related to the change in position: **Distance** and **Displacement**.

### Distance:
- **Definition:** The total path length covered during a journey.

### Displacement:
- **Definition:** The path length from the final position of the particle to the origin \(O\).

---

## Example

Consider the following figure:


----|B-------|0--------|A-------

We have an origin O. Measurements to the right of O are taken as a positive while to the left are taken as negative. Suppose a person who starts from origin O reaches point A,

Distance = OA

Displacement = OA

Now he turns and reaches point B,

Distance = OA + AB

Displacement = -OB

As we can see, displacement is negative since it is measured to the left of the origin. We can infer from the above example that distance is always positive while displacement can either be positive or negative.

# Speed and Velocity

These terms are used to describe the **rate of change of position**. While both speed and velocity measure how fast an object moves, they differ in how they treat direction.

- **Speed** is the rate of change in distance, a **scalar quantity** that only considers magnitude.
- **Velocity** is the rate of change in displacement, a **vector quantity** that considers both magnitude and direction.

---

## Key Differences

### Speed:
- Speed measures the **rate of change in distance** traveled.
- It is a **scalar quantity**, meaning it has no directionâ€”only magnitude.
- Speed is always **positive**, as distance cannot be negative.
  
**Formula:**
\[
\text{Speed} = \frac{\text{Distance}}{\text{Time}}
\]

### Velocity:
- Velocity measures the **rate of change in displacement**.
- It is a **vector quantity**, meaning it has both magnitude and direction.
- Velocity can be **positive** or **negative**, depending on the direction of the displacement.

**Formula:**
\[
\text{Velocity} = \frac{\Delta s}{\Delta t}
\]
Where:
- \(\Delta s\) = Displacement (change in position)
- \(\Delta t\) = Time interval

---

## Vector vs. Scalar

### Scalar Quantities:
- Scalars have **magnitude only**.
- Examples: Distance, Speed, Mass, Energy.

### Vector Quantities:
- Vectors have **magnitude and direction**.
- Examples: Displacement, Velocity, Force, Acceleration.

In the context of motion:
- **Speed** is scalar, meaning it tells you "how fast" something is moving without considering direction.
- **Velocity** is vector, meaning it tells you "how fast" and "in which direction."

---

## Example:

1. A car moves 50 meters north in 5 seconds:
   - Speed = \( \frac{50}{5} = 10 \, m/s \)
   - Velocity = \( \frac{50}{5} = 10 \, m/s \) (northward)

2. The same car then moves 30 meters south in 3 seconds:
   - Speed = \( \frac{30}{3} = 10 \, m/s \)
   - Velocity = \( \frac{-30}{3} = -10 \, m/s \) (southward, hence negative)

---

## Practical Applications:
- **Speed** is commonly used in daily life, like measuring how fast a car or a runner moves.
- **Velocity** is crucial in physics, where direction is essential, such as in calculating forces, projectiles, or motion analysis.

---

## Graphical Representation:

1. **Speed:**
   - Represented as a positive value on a graph.
   - Distance vs. Time graph has a positive slope.

2. **Velocity:**
   - Can be represented as positive or negative on a graph.
   - Displacement vs. Time graph can have positive and negative slopes.

---

## Conclusion:

- Use **speed** when direction is irrelevant and only the magnitude of motion matters.
- Use **velocity** when both magnitude and direction are important.
